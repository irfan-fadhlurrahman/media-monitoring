```python
import pytz
import time
import random
import pandas as pd
from datetime import datetime, timedelta
from GoogleNews import GoogleNews

def get_current_date(days_diff: int = 0):
    return (datetime.now(tz=pytz.timezone('Asia/Jakarta')) - timedelta(days=days_diff)) \
        .replace(tzinfo=None, microsecond=0)

def extract_google_news(keyword, selected_date):
    start = selected_date.strftime('%m/%d/%Y')
    end = (selected_date + timedelta(days=1)).strftime('%m/%d/%Y')

    googlenews = GoogleNews(lang='id')
    googlenews.set_time_range(start, end)
    googlenews.set_encode('utf-8')
    googlenews.get_news(keyword)

    df = pd.DataFrame(googlenews.results(sort=True))
    if df.empty:
        print(selected_date, 'FAILED')
        return pd.DataFrame()

    df['date_created'] = get_current_date()
    df['keyword'] = keyword

    googlenews.clear()
    del googlenews

    print(selected_date, 'SUCCESS', len(df), 'rows')
    return df

def load_to_drive(df_list, keyword):
    df_list = [df for df in df_list if not df.empty]
    if df_list:
        df = (
            pd.concat(df_list, ignore_index=True).reset_index(drop=True)
            .sort_values(by=['date_created'], ascending=False)
            .drop_duplicates(subset=['title', 'media'], keep='first')
            .reset_index(drop=True)
        )
        filename = f"/content/drive/MyDrive/News/{keyword}_{get_current_date().strftime('%Y%m%d')}.parquet"
        df.to_parquet(filename, index=False)
        print(f"Saved to {filename}")
        print(f"Total rows: {len(df)}\n")

def run(keyword, start_date, end_date):
    df_list = []
    for selected_date in pd.date_range(start_date, end_date, freq='D')[::-1]:
        df = extract_google_news(keyword=keyword, selected_date=selected_date)
        if len(df) > 0:
            df_list.append(df)
            tmp = load_to_drive(df_list=df_list, keyword=keyword)
        time.sleep(2)


# Temporary
import warnings
warnings.filterwarnings("ignore")

# Starts here
# keyword_list = [
#     'pajak motor',
#     'rupiah dolar as',
# ]

INVESTASI = [
    'bursa',
    'cryptocurrency',
    'danantara',
    'diversifikasi pasar',
    'harga emas',
    'harga properti',
    'harga saham',
    'ihsg',
    'investasi',
    'investasi syariah',
    'kripto',
    'obligasi',
    'obligasi syariah',
    'ori',
    'pasar modal',
    'reksadana',
    'resesi',
    'sbn',
    'yield',
]
PERTUMBUHAN_EKONOMI = [
    'anggaran kementerian',
    'anggaran negara',
    'apbn',
    'apbn perubahan',
    'bangkrut',
    'belanja kementerian',
    'belanja masyarakat',
    'belanja negara',
    'belanja pemerintah',
    'biaya pendidikan',
    'biaya sekolah',
    'daya beli masyarakat',
    'defisit anggaran',
    'ekonomi dunia',
    'ekonomi global',
    'ekonomi tumbuh',
    'gaji ke-13',
    'gaji naik',
    'gaji pegawai negeri',
    'gaji pns',
    'impor barang modal',
    'industri manufaktur',
    'industri otomotif',
    'industri tekstil',
    'investasi asing ',
    'investasi dalam negeri',
    'kenaikan gaji',
    'kenaikan pajak',
    'keyakinan konsumen',
    'konsumsi masyarakat',
    'korupsi',
    # 'lowongan kerja',
    # 'pailit',
    # 'pajak naik',
    # 'penanaman modal dalam negeri',
    # 'pendapatan masyarakat',
    # 'penerimaan pajak',
    # 'pengangguran',
    # 'pengangguran terbuka',
    # 'perbankan',
    # 'perdagangan',
    # 'perkebunan',
    # 'pertanian',
    # 'pertumbuhan ekonomi',
    # 'pmdn',
    # 'ppn naik',
    # 'proyeksi ekonomi',
    # 'rapbn',
    # 'resesi ekonomi',
    # 'suku bunga bank sentral',
    # 'thr',
    # 'tunjangan hari raya',
    # 'uang beredar',
    # 'umk',
    # 'umr',
    # 'upah',
    # 'utang pemerintah'
]
random.shuffle(PERTUMBUHAN_EKONOMI)
keyword_list = INVESTASI + PERTUMBUHAN_EKONOMI
start_date = get_current_date(365)
end_date = get_current_date()

for keyword in keyword_list:
    run(keyword=keyword, start_date=start_date, end_date=end_date)
```

```python
from minio import Minio
from io import BytesIO
from google.colab import userdata
from datetime import datetime, timedelta

STORAGE_OPTIONS = {
    "key": userdata.get('MINIO_ACCESS_KEY'),
    "secret": userdata.get('MINIO_SECRET_KEY'),
    "use_ssl": False,
    "client_kwargs": {
        "endpoint_url": f"http://{userdata.get('MINIO_ENDPOINT')}"
    }
}

def save_to_minio(df, object_name):
    client = Minio(
        userdata.get('MINIO_ENDPOINT'),
        access_key=userdata.get('MINIO_ACCESS_KEY'),
        secret_key=userdata.get('MINIO_SECRET_KEY'),
        secure=False,
    )
    buffer = BytesIO()
    df.to_parquet(buffer, index=False)
    buffer.seek(0)
    client.put_object(
        bucket_name=userdata.get('MINIO_BUCKET_NAME'),
        object_name=object_name,
        data=buffer,
        length=buffer.getbuffer().nbytes,
    )
    print(f"Saved to s3://{userdata.get('MINIO_BUCKET_NAME')}/{object_name}")

save_to_minio(df, object_name="tmp/pertumbuhan_ekonomi_20250415.parquet")
```
